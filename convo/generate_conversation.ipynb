````python
import openai

# Retrieve API key from environment variable
openai.api_key = ""

def inference_openai(model: str, prompt: str, 
                     system_message: str = "given the text, some parts are merged together due to pdf parsing , i need you to reply and use word by word match and only reorganize the text to have continious paragraph"
                     ) -> str:
    try:
        response = openai.ChatCompletion.create(
            model=model,
            messages=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt}],
        )
        return response['choices'][0]['message']['content']
    except Exception as e:
        return f"Error: {str(e)}"
````

````python
def run_llm(system, prompt):
        answer = inference_openai("gpt-4o-mini", prompt, system)
        print(answer)
        return answer


def person_to_string(person):
  # return the name of the person, the department they are in, and their traits in a single string
  return f"{person['name']} is in the {people[person['name']]} department and has the following traits: {person['traits']}"

def task_relationship_score(
  person1, person2, tasks1, tasks2
):
  # use LLM to calculate the relationship score, give the llm the two people's names and their tasks and their personalities
  
  # generate the prompt
  prompt = f"""Given the tasks and personalities of {person1} and {person2}:
    person1: {person_to_string(person1)}, tasks: {tasks1},
    person2: {person_to_string(person2)}, tasks: {tasks2},
    What is the likelihood that they will have that they will talk about the tasks they are working on?
    return a score between 0 and 1, and nothing more than that. return only the number."""
    
  system = "Calculate the relationship score between two people based on their tasks and personalities and department and determine the likelihood that they will talk about the tasks they are working on."
  
  # get the answer from the LLM
  answer = run_llm(system, prompt)
    
  try:
    answer = float(answer)
  except:
    print("Error: Could not convert answer to float", person1, person2, answer)
    answer = 0.0
  
  return float(answer)

def generate_messages(
  person1, person2, tasks1, tasks2
):
  # generate messages between two people based on their tasks and personalities return a array of messages containing the messages ( timestamps are required , the sender and the message )
  
  # use LLM to generate the messages, give the llm the two people's names and their tasks and their personalities
  
  # generate the prompt  
  prompt = f"""
    Create a realistic chat conversation between these two professionals:

    IT IS IMPERATIVE THAT YOU FOLLOW THE INSTRUCTIONS BELOW TO ENSURE THE CONVERSATION IS REALISTIC AND USEFUL.
    AND DON T FORGET TO USE THE PERSONALITIES OF THE PEOPLE INVOLVED.
    AND MENTION THE NAMES OF THE PEOPLE IN THE CONVERSATION.

    Person 1: {person_to_string(person1)}
    task: {tasks1}
    
    Person 2: {person_to_string(person2)}
    task: {tasks2}

    Requirements for the conversation:
    1. Make the dialogue reflect each person's unique communication style:
       - Use appropriate technical jargon based on their expertise level
       - Mirror their personality traits in word choice and tone
       - Include realistic reactions based on their background
    
    2. Include specific technical details about their tasks:
       - Reference actual project milestones
       - Mention concrete technical challenges
       - Use relevant industry terminology
    
    3. Add human elements:
       - Include small talk where natural
       - Show personality quirks
       - Include typos or casual language if it fits their personality
    
    4. Make timestamps realistic:
       - respect the dates of the tasks
       - Account for response delays
       - Consider time zones if relevant
       - Show realistic work patterns (breaks, meetings, etc.)
       - Include timestamps for each message and make them consistent with the conversation flow
       
    5. You might generate aggressive or passive-aggressive messages if it fits the personalities of the people involved, use slur words if necessary.
    
    7. use both people's names in the conversation. unless it is a ghosting but only do that if it fits the personalities of the people involved.

    1. Message Structure:
       - Include specific technical terms, metrics, and project details
       - Reference actual code, systems, or architectural components
       - Vary message lengths naturally
       
        2. Technical Depth:
       - Discuss concrete implementation details
       - Reference specific technical challenges and solutions
       - Include data points and metrics where relevant
       - Mention actual tools, frameworks, and methodologies
    
    3. Professional Tone:
       - Maintain professional but natural language
       - Limit emojis to 1-2 per conversation, not per message
       - Focus on technical substance over management clichés
       - Include occasional brief acknowledgments

    4. Conversation Flow:
       - Allow natural gaps between messages (1-60 minutes) , more in some cases
       - Include relevant follow-ups to technical points
       - Mix synchronous and asynchronous communication patterns
       - Show realistic problem-solving progression
    
    5. try to make the conversation a bit lenghty to cover the tasks and the personalities of the people involved.

    Example of desired technical detail level:
    "I've analyzed the query performance logs from last night's batch processing - seeing consistent 2.3s latency spikes during the JSON aggregation phase. Might need to refactor the nested loops in the data transformation layer. Here's the problematic section: [technical details]. Thoughts on implementing a streaming approach instead?"



    Format the output as a JSON array of messages, where each message has:
    - timestamp: ISO 8601 format
    - name: full name of the sender
    - message: the actual message content
    """

  system = """
    Role: Conversation Generator
    Task: Generate authentic workplace messages between two professionals
    Output: JSON array of timestamped messages
    Focus: Maintain consistent personality traits and technical accuracy while keeping dialogue natural
    """
  
  answer = run_llm(system, prompt)
  
  # return the answer as a string
  return answer

  
````
